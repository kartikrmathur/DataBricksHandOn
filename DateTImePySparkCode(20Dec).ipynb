```
import datetime as dt
import pandas as pd
import pyspark
```

```
#defining Function
def getDate(date_text, digit_no):
  if date_text is None or date_text.strip() == "":
    return ""
    date_text = date_text.strip()
#if only 2 digits default it to year after validate it is numerical
  if len(date_text) == 2 and date_text.isdigit():
    date_text = date_text + "/01/01" # yy/mm/dd

#list of formats

  formats = [
"%m/%d/%Y", # mm/dd/yyyy
"%Y/%m/%d", # yyyy/mm/dd
"%d/%m/%Y", # dd/mm/yyyy
"%y/%m/%d", # yy/mm/dd
"%Y/%m", # yyyy/mm
"%m-%d-%Y", # mm-dd-yyyy
"%Y-%m-%d", # yyyy-mm-dd
"%d-%m-%Y", # dd-mm-yyyy
"%m-%d-%y", # mm-dd-yy
"%y-%m-%d", # yy-mm-dd
"%b %d %Y", # Jan 08 2017
"%b %d %Y %H:%M:%S", # Jan 08 2017 00:00:00
"%b,%d,%Y %H:%M:%S", # Jan,08,2017 00:00:00
"%b, %d, %Y %H:%M:%S", # Jan, 08, 2017 00:00:00
"%m-%d-%Y %H:%M:%S", # mm-dd-yyyy 00:00:00
"%Y%m", # yyyymm
"%Y%m%d", # yyyymmdd
"%d%b%Y", # 08JAN2017
"%d%b%Y:%H:%M:%S", # 08JAN2017:00:00:00
"%b %d %Y %H:%M%p", # JAN 08 2017 00:00AM
"%b-%y", # MAR-98
"%d-%b-%Y", # 02-MAR-1998
"%d-%b-%y", # 02-MAR-98
  ]
#try Catch Block
  for fm in formats:
#exception Handling  
    try:
      return dt.datetime.strptime(date_text, fm).strftime("%Y%m%d")[:digit_no]
    except ValueError: 
      pass 
      return ""
```
